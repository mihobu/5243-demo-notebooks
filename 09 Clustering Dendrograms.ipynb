This notebook contains code for generating dendrograms for examples in section 7.3 of the textbook (Tan et al, 2nd edition) .. (Chapter 8 of the first edition)

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics.pairwise import pairwise_distances
```

```python
np.set_printoptions(precision=3, linewidth=125)
```

```python
X = np.array([
    [0.4005, 0.5306],
    [0.2148, 0.3854],
    [0.3457, 0.3156],
    [0.2652, 0.1875],
    [0.0789, 0.4139],
    [0.4548, 0.3022]
])
labels = list(range(1,7))
```

```python
fig = plt.figure(figsize=(6, 6), dpi=100)
ax = fig.add_subplot(1, 1, 1)

# Label the cluster centers
for i, c in enumerate(X):
    cl = '${}$'.format(i+1)
    ax.scatter(c[0], c[1], marker='o', c="blue", alpha=0.5, s=200, edgecolor='')
    ax.scatter(c[0], c[1], marker=cl, alpha=1, s=50, c='white', edgecolor='')

plt.show()
```

```python
points = pairwise_distances(X)
print(points)
```

## Example 8.4: Single Link (MIN)

```python
Z = linkage(X, method='single', metric='euclidean')
```

```python
def pointName(i):
    return(['p1', 'p2', 'p3', 'p4', 'p5', 'p6'][i])
```

```python
# calculate full dendrogram
plt.figure(figsize=(10, 6), dpi=200)
plt.title('Example 8.4 (Single Link)')
plt.xlabel('Point')
plt.ylabel('Proximity')
dendrogram(
    Z,
    leaf_rotation=0.,  # rotates the x axis labels
    leaf_font_size=10.,  # font size for the x axis labels
    leaf_label_func=pointName   
)
#plt.ylim(0, 1)
plt.show()
```
