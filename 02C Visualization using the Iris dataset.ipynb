# Visualization
In this notebook, we'll learn how to do some basic visualizations using the
[Pyplot](https://matplotlib.org/api/pyplot_summary.html) modules of
[Mtplotlib](https://matplotlib.org/).
We'll be using the famous [Iris dataset](https://archive.ics.uci.edu/ml/datasets/Iris).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# If you don't already have a copy of the Iris data file, you can download it here.
IRISURL = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
import urllib
urllib.request.urlretrieve (IRISURL, "data/iris.csv")
```

```python
iris_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris = pd.read_csv('data/iris.csv', names=iris_columns)
```

## Histogram

```python
# By default, the hist() method will display histograms for all numeric attributes.
iris.plot.hist()
```

```python
# To show a histogram on a single attribute, call using the attribute itself.
iris['sepal_width'].plot.hist()
```

```python
# Customize the number of bins
iris['sepal_width'].plot.hist(bins=20)
```

## Kernel density plot

Get an estimate, based on the available data, of the population distribution.

```python
# As before, show plots for all attributes
iris.plot.kde()
```

## Box plot

Box and whisker plot using Tukey's fences ($k=1.5$)

```python
# We can look at the distributions of each variable.
iris.boxplot()
```

```python
# If we want to limit the variables, we can do so
iris.boxplot(column=['sepal_length', 'sepal_width'])
```

```python
# Single variable by species
iris.boxplot(column='petal_length', by='class')
```

## Scatter plot

```python
# Simple scatter plot of two variables
iris.plot.scatter('sepal_length', 'sepal_width')
```

```python
# It would be nice to show the species in different colors.
groups = iris.groupby('class')
fig, ax = plt.subplots()
for name, group in groups:
    ax.plot(group.petal_length, group.petal_width, marker='o', linestyle='', label=name)
ax.legend()
plt.show()
```

## Scatter plot matrix

```python
pd.plotting.scatter_matrix(iris, figsize=(10, 10))
plt.show()
```

## Parallel coordinates chart
There are many [colormaps](https://matplotlib.org/users/colormaps.html) to choose from.

```python
pd.plotting.parallel_coordinates(iris, 'class', colormap='Set1')
```

## Controlling plot size

```python
# If you'd like the plot to be larger, you do something like this before your plot command:
plt.figure(num=None, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')
plt.clf()
pd.plotting.parallel_coordinates(iris, 'class', colormap='viridis')
plt.show() # Suppress the "<matplotlib.axes ....>" output
```
